### Importación de librerias

```python
import pandas as pd
import numpy as np
```

## Extraer datos correspondientes para poder realizar el grafico

```python
df = pd.read_csv('data/casos_nuevos_acumulados_por_fecha.csv', index_col = 0) 
casosN= df[df.Serie != 'Casos acumulados']
```

## Creación del grafico con sus datos correspondientes

```python
ejeX=np.arange(len(casosN.index),step= 7)
xlabels=list(casosN.index)
casosN.plot(figsize=(15,10),xlabel='Fecha',ylabel='Casos Nuevos',
            title ='Cantidad casos nuevos por fecha (desde el 03/03/2020 hasta el 05/09/2021',xticks=ejeX,rot=90)
```

## Extraer datos correspondientes para poder realizar el grafico

```python
casosA= df[df.Serie != 'Casos nuevos']
```

## Creación del grafico con sus datos correspondientes

```python
ejeX=np.arange(len(casosN.index),step= 7)
xlabels=list(casosN.index)
casosA.plot(figsize=(15,10),xlabel='Fecha',ylabel='Casos Acumulados',
            title ="Cantidad casos acumulados por fecha (desde el 03/03/2020 hasta el 05/09/2021)",xticks=ejeX,rot=90)
```

## Comparación de datos entre casos nuevos e ingresos UCI

```python
df = pd.read_csv('data/Ingresos_UCI.csv', index_col = 0) 
mezcla= df.T.merge(casosN,left_index=True,
    right_index=True)
mezcla.drop('Serie', axis=1, inplace= True)
ejeX=np.arange(len(casosN.index),step= 7)
xlabels=list(casosN.index)
mezcla.plot(figsize=(15,10),xlabel='Fecha',ylabel='Casos Nuevos e Ingresos UCI',
            title ='Cantidad casos nuevos e ingresos UCI por fecha (desde el 03/03/2020 hasta el 05/09/2021)',xticks=ejeX,rot=90)
```
