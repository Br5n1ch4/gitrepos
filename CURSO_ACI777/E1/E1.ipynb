# Importación de librerias

```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from array import array
```

## Extraer datos correspondientes para poder realizar el grafico

```python
fechaN=[]
casoN=[]
with open('data/casos_nuevos_acumulados_por_fecha.csv') as cnapf:
    linea = csv.reader(cnapf, delimiter=',')
    cont=1
    contador=1
    for columna in linea:
        if columna[1]=="Casos nuevos":
            fechaN.append(columna[0])
            casoN.append(float(columna[2]))      
```

## Creación del grafico con sus datos correspondientes

```python
fig, ax =plt.subplots(figsize=(20,10))
plt.rc('font', size=25)
ejeX=np.arange(len(fechaN))
barraN=ax.plot(ejeX,casoN,label='Casos nuevos',color="orange")
ax.set_title("Cantidad casos nuevos por fecha (desde el 03/03/2020 hasta el 05/09/2021)", 
             fontdict = {'fontsize':30, 'fontweight':'bold', 'color':'tab:green'})
ax.set_xlabel("Fecha", fontdict = {'fontsize':30, 'fontweight':'bold', 'color':'tab:green'})
ax.set_ylabel("Casos Nuevos", fontdict = {'fontsize':30, 'fontweight':'bold', 'color':'tab:green'})
ax.legend()
ax.set_xticklabels(fechaN, rotation=90)
```
